```python
import numpy as np
import pandas as pd
```

```python
맛집_df = pd.read_csv('/content/drive/MyDrive/식당소개.csv', encoding = 'CP949',  on_bad_lines='skip')
```

```python
맛집_df = 맛집_df.loc[ : , ['conts_ttl', 'addr', 'map_lat', 'map_lot', 'conts_sumry', 'bsn_hr_guid', 'hldy_guid']]
```

```python
맛집_df.rename(columns =
    {
        'conts_ttl' : '식당명',
        'addr' : '도로명 주소',
        'map_lat' : 'lat',
        'map_lot' : 'lot',
        'conts_sumry' : '식당 설명',
        'bsn_hr_guid' : '영업 시간',
        'hldy_guid' : '휴일 정보'
    }, inplace=True
)
```

```python
맛집_df.insert(1, '행정구역', 맛집_df['도로명 주소'].str.extract(r'(?:인천(?:광역시|시)?\s*)?([가-힣]+[구군])'))
```

```python
맛집_df.dropna(subset=['행정구역'], inplace=True)
```

```python
맛집_df['행정구역'].unique()
```

```python
맛집_df['식당명'] = 맛집_df['식당명'].astype(str)
맛집_df['행정구역'] = 맛집_df['행정구역'].astype(str)
맛집_df['도로명 주소'] = 맛집_df['도로명 주소'].astype(str)
맛집_df['식당 설명'] = 맛집_df['식당 설명'].astype(str)
맛집_df['영업 시간'] = 맛집_df['영업 시간'].astype(str)
맛집_df['휴일 정보'] = 맛집_df['휴일 정보'].astype(str)
```

```python
맛집_df.to_csv('인천광역시 식당 현황.csv', index=False, encoding='euc-kr')
```
