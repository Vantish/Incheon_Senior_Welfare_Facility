```python
# -----------------------------
# 📍 인천 노인복지시설 지도 시각화 & 가장 가까운 시설 찾기
# -----------------------------

import numpy as np
import folium
from math import radians, cos, sin, asin, sqrt
import re

# 1️⃣ 엑셀 파일 불러오기
df = pd.read_csv('노인복지시설_가공.csv', encoding='utf-8-sig')

# 2️⃣ 위도·경도 컬럼 찾기
lat_col, lon_col = '위도', '경도'  # 엑셀에 '위도', '경도' 열이 있다면 그대로 사용
address_col = '주소'

# 3️⃣ 주소에서 '구' 추출
def extract_gu(addr):
    if not isinstance(addr, str):
        return np.nan
    m = re.search(r'([가-힣]+구)', addr)
    return m.group(1) if m else np.nan

df['구'] = df[address_col].apply(extract_gu)

# 4️⃣ 좌표 결측치 제거
df = df.dropna(subset=[lat_col, lon_col])

# 5️⃣ 지도 생성 (인천 중심 좌표)
center_lat, center_lon = df[lat_col].mean(), df[lon_col].mean()
m = folium.Map(location=[center_lat, center_lon], zoom_start=11)

# 6️⃣ 구별로 색상 다르게 표시
palette = [
    'red','blue','green','purple','orange','darkred','lightred','beige','darkblue',
    'darkgreen','cadetblue','darkpurple','white','pink','lightblue','lightgreen','gray','black'
]
districts = df['구'].fillna('unknown').unique()
color_map = {d: palette[i % len(palette)] for i, d in enumerate(sorted(districts))}

for _, row in df.iterrows():
    lat, lon = row[lat_col], row[lon_col]
    name = row.get('시설명', '이름없음')
    gu = row.get('구', '')
    addr = row.get('주소', '')
    popup = folium.Popup(f"<b>{name}</b><br>{gu}<br>{addr}", max_width=300)
    folium.CircleMarker(
        location=[lat, lon],
        radius=5,
        color=color_map.get(gu, 'blue'),
        fill=True,
        fill_opacity=0.8,
        popup=popup
    ).add_to(m)

# 7️⃣ 기본 지도 저장
m.save('인천노인복지시설_지도.html')
print("✅ 지도 저장 완료: 인천노인복지시설_지도.html")

# 8️⃣ 거리 계산 함수 (하버사인 공식)
def haversine(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1)*cos(lat2)*sin(dlon/2)**2
    return 6371 * 2 * asin(sqrt(a))  # km 단위

# 9️⃣ 사용자 위치에서 가장 가까운 시설 찾기
def nearest_facility(user_lat, user_lon):
    df['거리(km)'] = df.apply(lambda r: haversine(user_lat, user_lon, r[lat_col], r[lon_col]), axis=1)
    nearest = df.loc[df['거리(km)'].idxmin()]

    info = {
        '시설명': nearest['시설명'],
        '주소': nearest['주소'],
        '구': nearest['구'],
        '거리(km)': round(nearest['거리(km)'], 2),
        '위도': nearest[lat_col],
        '경도': nearest[lon_col]
    }

    # 지도 표시
    map_user = folium.Map(location=[user_lat, user_lon], zoom_start=13)
    # 사용자 위치
    folium.Marker([user_lat, user_lon], icon=folium.Icon(color='red', icon='home'),
                  popup='내 위치').add_to(map_user)
    # 가까운 시설
    folium.Marker([info['위도'], info['경도']],
                  popup=f"{info['시설명']} ({info['거리(km)']}km)",
                  icon=folium.Icon(color='blue', icon='info-sign')).add_to(map_user)
    # 거리 연결선
    folium.PolyLine([[user_lat, user_lon], [info['위도'], info['경도']]], color='gray', weight=3).add_to(map_user)
    
    map_user.save('내위치_가까운시설.html')
    print(f"✅ '{info['시설명']}' (거리: {info['거리(km)']} km)")
    print("지도 저장: 내위치_가까운시설.html")
    return info

# 10️⃣ 사용 예시
# 내 위치를 위도, 경도로 입력
# 예: nearest_facility(37.456, 126.705)
```
